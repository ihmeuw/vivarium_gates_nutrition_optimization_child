```python
# Locations [Sub-Saharan Africa, South Asia, Ethiopia, India, Nigeria]
output_dir = '/share/costeffectiveness/auxiliary_data/GBD_2019/01_original_data/population_attributable_fraction/risk_factor/lbwsg/data/nigeria/'
artifact_path = '/share/costeffectiveness/artifacts/vivarium_gates_nutrition_optimization_child/nigeria.hdf'
config = '/ihme/homes/albrja/repos/sim_sci/vivarium_gates_nutrition_optimization_child/src/vivarium_gates_nutrition_optimization_child/data/lbwsg_paf_sub-saharan_africa.yaml'
draw = 10
seed = 15
```

```python
def runner(draw: int) -> None:
    return lbwsg_paf_calculator.write_pafs_to_hdf(config, artifact_path, output_dir, draw, 0)
```

```python
def run_parallel(n_draws: int) -> List[float]:
    draws = list(range(n_draws))
    
    with multiprocessing.Pool(30) as pool:
        results = list(tqdm.tqdm(
            pool.imap(runner, draws), 
            total=len(draws),
        ))
    return results
```
